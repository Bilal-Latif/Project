```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

csv_file_path = "homeprices.csv"  
df_loaded = pd.read_csv(csv_file_path)

print("Dataset loaded from homeprices.csv:")
print(df_loaded)

X = df_loaded[['area']]  
y = df_loaded['price'] 

model = LinearRegression()
model.fit(X, y)

predicted_prices = model.predict(X)

df_loaded['predicted_price'] = predicted_prices

print("\nUpdated dataset with predicted prices:")
print(df_loaded)

plt.scatter(df_loaded['area'], df_loaded['price'], color='blue', label='Actual Prices')
plt.plot(df_loaded['area'], df_loaded['predicted_price'], color='red', label='Regression Line')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Home Prices Prediction')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

csv_file_path = "homeprices.csv"
df_loaded = pd.read_csv(csv_file_path)

print("Dataset loaded from homeprices.csv:")
print(df_loaded)

X = df_loaded[['area']]  
y = df_loaded['price'] 

model = LinearRegression()
model.fit(X, y)
predicted_prices = model.predict(X)
df_loaded['predicted_price'] = predicted_prices
print("\nUpdated dataset with predicted prices:")
print(df_loaded)

areas_to_predict = [[5000], [8000], [9000]]
predicted_prices_for_areas = model.predict(areas_to_predict)
for area, price in zip(areas_to_predict, predicted_prices_for_areas):
    print(f"The predicted price of a home with area {area[0]} sq ft is ${price:.2f}")

plt.scatter(df_loaded['area'], df_loaded['price'], color='blue', label='Actual Prices')
plt.plot(df_loaded['area'], df_loaded['predicted_price'], color='red', label='Regression Line')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Home Prices Prediction')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

csv_file_path = "shopprices.csv"
df_loaded = pd.read_csv(csv_file_path)

print("Dataset loaded from shopprices.csv:")
print(df_loaded)

df_loaded['Price ($)'] = df_loaded['Price ($)'].replace({',': ''}, regex=True).astype(float)

X = df_loaded[['ShopArea (sq ft)']] 
y = df_loaded['Price ($)']  

model = LinearRegression()
model.fit(X, y)

predicted_prices = model.predict(X)

df_loaded['predicted_price'] = predicted_prices

print("\nUpdated dataset with predicted prices:")
print(df_loaded)


plt.scatter(df_loaded['ShopArea (sq ft)'], df_loaded['Price ($)'], color='blue', label='Actual Prices')
plt.plot(df_loaded['ShopArea (sq ft)'], df_loaded['predicted_price'], color='red', label='Regression Line')
plt.xlabel('Shop Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Shop Prices Prediction')
plt.legend()
plt.show()
```
